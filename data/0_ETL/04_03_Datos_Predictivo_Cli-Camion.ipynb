### Transformo algunas cosas para que se vea Tabla_1 

```python
#  Ubica Camion 
merged_df['Ubicacion_Camion'] = merged_df.apply(
    lambda row: f"https://www.google.com/maps?q={row['camion_x']},{row['camion_y']}", axis=1
)
```

```python
#  Ubicacion_Cliente con  Google Maps 
merged_df['Ubicacion_Cliente'] = merged_df.apply(
    lambda row: f"https://www.google.com/maps?q={row['LATITUD']},{row['LONGITUD']}", axis=1
)
```

```python
# Recupero dia y hora 
merged_df['Dia_camion_estac'] = merged_df['Tiempo_de_Inicio'].dt.date
merged_df['Hora_camion_desde'] = merged_df['Tiempo_de_Inicio'].dt.time
```

```python
# Tabla Final 
Tabla_Final = merged_df[['Numero_de_placa', 'Dia_camion_estac', 'Hora_camion_desde', 
                         'Ubicacion_Camion', 'Duracion', 'CODIGO', 'NOMCLI', 'Ubicacion_Cliente']]
```

```python
# Formatear fecha 
Tabla_Final['Dia_camion_estac'] = pd.to_datetime(Tabla_Final['Dia_camion_estac']).dt.strftime('%d-%m-%Y')
```

```python

# Formatear hora 
#Tabla_Final['Dia_camion_estac'] = pd.to_datetime(Tabla_Final['Dia_camion_estac']).dt.strftime('%d-%m-%Y')

Tabla_Final['Hora_camion_desde'] = Tabla_Final['Hora_camion_desde'].astype(str).str[:5]  # Mantiene solo 'HH_MM'

# Mostrar el resultado
print(Tabla_Final[['Dia_camion_estac', 'Hora_camion_desde']])
```

```python

# Convertir las columnas a string en caso de que estén en formato bytes
Tabla_Final['Ubicacion_Camion'] = Tabla_Final['Ubicacion_Camion'].astype(str)
Tabla_Final['Ubicacion_Cliente'] = Tabla_Final['Ubicacion_Cliente'].astype(str)

# Definir la función para extraer coordenadas de Ubicacion_Camion
def extract_coordinates_from_camion(url):
    match = re.search(r'(-?\d+\.\d+),(-?\d+\.\d+)', url)
    if match:
        return float(match.group(1)), float(match.group(2))
    else:
        return None, None

# Definir la función para extraer coordenadas de Ubicacion_Cliente
def extract_coordinates_from_cliente(url):
    # Intentar encontrar el formato con o sin paréntesis, permitiendo espacios en blanco
    match = re.search(r'\(?\s*(-?\d+\.\d+)\s*,\s*(-?\d+\.\d+)\s*\)?', url)
    if match:
        return float(match.group(1)), float(match.group(2))
    else:
        return None, None

# Aplicar las funciones para extraer las coordenadas en nuevas columnas
Tabla_Final[['lat_camion', 'lon_camion']] = Tabla_Final['Ubicacion_Camion'].apply(
    lambda x: pd.Series(extract_coordinates_from_camion(x))
)
Tabla_Final[['lat_cliente', 'lon_cliente']] = Tabla_Final['Ubicacion_Cliente'].apply(
    lambda x: pd.Series(extract_coordinates_from_cliente(x))
)

# Calcular la distancia Haversine
def haversine(lat1, lon1, lat2, lon2):
    # Verificar que ninguna coordenada sea None
    if None in [lat1, lon1, lat2, lon2]:
        return np.nan  # Devuelve NaN si falta alguna coordenada
    R = 6371000  # Radio de la Tierra en metros
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    a = np.sin(delta_phi / 2) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    return R * c  # Distancia en metros

# Calcular la distancia en metros y agregarla en una nueva columna
Tabla_Final['distancia_en_Metro'] = Tabla_Final.apply(
    lambda row: haversine(row['lat_camion'], row['lon_camion'], row['lat_cliente'], row['lon_cliente']), axis=1
)

# Revisar los primeros valores para confirmar el cálculo
print(Tabla_Final[['lat_camion', 'lon_camion', 'lat_cliente', 'lon_cliente', 'distancia_en_Metro']].head())

```

```python
# Redondear 'distancia_en_Metro' a dos decimales
Tabla_Final['distancia_en_Metro'] = Tabla_Final['distancia_en_Metro'].round(2)

# Eliminar las columnas de coordenadas solo si existen
columns_to_drop = ['lat_camion', 'lon_camion', 'lat_cliente', 'lon_cliente']
Tabla_Final = Tabla_Final.drop(columns=[col for col in columns_to_drop if col in Tabla_Final.columns])
```
